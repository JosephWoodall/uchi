# Uchi Quickstart Guide

Welcome to **Uchi**! This notebook will walk you through the core components of the library: sequence prediction, tabular classification, and time series forecasting.

```python
# 1. Install uchi if you haven't already
!pip install uchi_python

from uchi import UniversalPredictor

# 2. Initialize a basic sequence predictor
predictor = UniversalPredictor(depth=3)

# 3. Train on a sequence of characters
sequence = "hello world! hello python! hello uchi!"
for char in sequence:
    predictor.observe(char)

# 4. Predict the next character given a prefix
prefix = "hell"
probabilities = predictor.predict(prefix)
print(f"Probabilities after '{prefix}':", probabilities)
```

## Tabular Data
Uchi is completely scikit-learn compatible!

```python
import numpy as np
from uchi import TabularPredictor
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Generate mock data
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Train tabular predictor
clf = TabularPredictor(max_depth=4)
clf.fit(X_train, y_train)

print("Accuracy:", clf.score(X_test, y_test))
```
